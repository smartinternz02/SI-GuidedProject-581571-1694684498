```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_crashes = sns.load_dataset("car_crashes")
```

PAIRPLOT

```python
sns.pairplot(car_crashes)
plt.title("Pairplot of Car Crashes Dataset")
plt.show()
```

INFERENCE:The pairplot provides scatterplots for all numerical variables in the dataset, allowing us to visually explore relationships between variables. For example, there seems to be a positive correlation between alcohol and total crashes.

HISTOGRAM OF ALCOHOL CONSUMPTION


```python
plt.figure(figsize=(8, 6))
sns.histplot(car_crashes['alcohol'], bins=15, kde=True)
plt.title("Histogram of Alcohol Consumption")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Frequency")
plt.show()
```

INFERENCE:The histogram shows the distribution of alcohol consumption. Most states have a relatively low alcohol consumption rate, with a peak around 0.1 to 0.15. There are a few outliers with very high alcohol consumption.

Boxplot of Speeding vs. Total Crashes

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='speeding', y='total', data=car_crashes)
plt.title("Boxplot of Speeding vs. Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
plt.show()
```

Inference: The boxplot illustrates the relationship between speeding and the total number of crashes. States with higher speeding rates tend to have a wider spread of total crashes, with some outliers indicating more crashes in those states.

Barplot of State vs. Total Crashes

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='abbrev', y='total', data=car_crashes, palette='Set3')
plt.title("Barplot of State vs. Total Crashes")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")
plt.xticks(rotation=45)
plt.show()
```

Regression Plot of Alcohol vs. Total Crashes

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='alcohol', y='total', data=car_crashes)
plt.title("Regression Plot of Alcohol vs. Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
plt.show()
```

 Inference: The regression plot shows a linear relationship between alcohol consumption and total crashes. As alcohol consumption increases, the total number of crashes also tends to increase.

Violinplot of Alcohol Consumption by State

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='abbrev', y='alcohol', data=car_crashes, palette='Pastel1')
plt.title("Violinplot of Alcohol Consumption by State (Abbreviation)")
plt.xlabel("State Abbreviation")
plt.ylabel("Alcohol Consumption")
plt.xticks(rotation=45)
plt.show()
```

Swarmplot of Speeding vs. Alcohol Consumption

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x='speeding', y='alcohol', data=car_crashes, palette='Set2')
plt.title("Swarmplot of Speeding vs. Alcohol Consumption")
plt.xlabel("Speeding")
plt.ylabel("Alcohol Consumption")
plt.show()
# Inference: The swarmplot reveals the relationship between speeding and alcohol consumption. It appears that states with higher speeding rates tend to have a wider range of alcohol consumption values.
```

Heatmap of Correlation Matrix

```python
correlation_matrix = car_crashes.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

# Inference: The heatmap visualizes the correlation between numerical variables in the dataset. Variables like alcohol and total crashes have a strong positive correlation, while speeding and total crashes have a weaker negative correlation.

Jointplot of Alcohol vs. Speeding

```python
sns.jointplot(x='alcohol', y='speeding', data=car_crashes, kind='scatter')
plt.suptitle("Jointplot of Alcohol vs. Speeding")
plt.show()
```

Inference: The jointplot provides a scatter plot of alcohol consumption vs. speeding, along with histograms of each variable. There doesn't seem to be a clear linear relationship between the two variables.
